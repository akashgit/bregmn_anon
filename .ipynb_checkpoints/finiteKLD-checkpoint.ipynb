```python
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
import numpy as np
import seaborn as sns; sns.set(color_codes=True)
import tensorflow_probability as tfp
tfd = tfp.distributions
import matplotlib.pyplot as plt
```

```python
mx = tf.Variable([0.,0.])
sigmax = tf.Variable([1., 1.])
sigx = tf.sqrt(tf.exp(sigmax))
p_x = tfd.Normal(loc=mx, scale=sigx)

m = tf.Variable([0.,0.])
sigma = tf.Variable([2., 2.])
sig = tf.sqrt(tf.exp(sigma))

q = tfd.Normal(loc=m, scale=sig)
samples = q.sample([32]) 

kld = tf.reduce_mean(tf.square((p_x.prob(samples)/q.prob(samples)) - 1))
x2 = tf.reduce_sum((p_x.prob(samples)/q.prob(samples)))
# kld = -tf.reduce_mean( q.log_prob(samples) - p_x.log_prob(samples) )
x3 = -tf.reduce_mean( tf.log(q.prob(samples) + 1e-6) - tf.log(p_x.prob(samples) + 1e-6) )

opt = tf.train.AdamOptimizer(learning_rate=1e-2)
train = opt.minimize(-kld, var_list=[sigmax,sigma])
init = tf.global_variables_initializer()
```

```python
batch_s = 5000
batch_sample = 500
# Start Training
# Start a new TF session
sess = tf.Session()

# Run the initializer
sess.run(init)

# Training
kld_store=[]
for i in range(0, 50001):
    
    m_,sig_,m__,sig__ = sess.run([m,sig,mx,sigx])
    # Train
    feed_dict = {}
    _, kl, x2_, x3_ = sess.run([train, kld, x2, x3],feed_dict=feed_dict)
    

    if i%5000==0:
        print(x3_)
        print(x2_)
        print(kl)
        print(m_,sig_)
        print(m__,sig__)
        kld_store.append(kl)
        

s2 = np.array([np.random.normal(m_[0], sig_[0],size=[batch_s]),np.random.normal(m_[1], sig_[1],size=[batch_s])]).T
ax2 = sns.kdeplot(s2[:,0],s2[:,1],cmap="Reds", shade=True, shade_lowest=False)

s1 = np.array([np.random.normal(m__[0], sig__[0],size=[batch_s]),np.random.normal(m__[1], sig__[1],size=[batch_s])]).T
ax1 = sns.kdeplot(s1[:,0],s1[:,1],cmap="Blues", shade=True, shade_lowest=False) 
```

```python
kld_store_save_ = kld_store
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(kld_store_save[:-1],label='via Ratio')
plt.plot(kld_store_save_[:-1],label='Direct')
plt.xlabel("Iterations x5000")
plt.ylabel("Pearson Divergence")
plt.legend(loc='upper left')
plt.show()
```
